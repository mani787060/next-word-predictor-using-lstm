```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index

# we can see that each word assigns with a unique number.
```

```python
# let's check our whole dataset on the basis of 'line_change' :-

for sentence in faqs.split('\n'):
    print(sentence)

# so these are our all sentences below    
```

```python
# Now, let's represent our whole sentence in numbers :-

for sentence in faqs.split('\n'):
    print(tokenizer.texts_to_sequences([sentence])[0])

# now our all sentences are converted into numericals    
```

```python
input_sequences = [] 

for sentence in faqs.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
# So now we have to make all these sentences of equal length
# that's why we add extra 'zeros' next to the numbers which have less length (as compared to the longest word) 
# for this first we have to find the lenght of the 'longest_word' from this whole dataset
  
max_len = max([len(x) for x in input_sequences])
```

```python
# now we use 'zero_padding'(i.e., adding extra zeros) 

from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre') 
```

```python
padded_input_sequences
```

```python
# now we have to make it as a 'supervised_learning' dataset :-

X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
# so this is a "MULTICLASS-CLASSIFICATION-TASK"

# that's why we have to Apply OHE on the 'y' for changing it into categorical form.
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes=283)     # 'num_classes' -> it is total no. of unique words(i.e., 282) present in our dataset.

# we give here 283 instead of 282, b/c OHE always starts from 0(zero),so this consider 281 as its last word (if we take 282).
```

```python
y
```

```python
y.shape  # i.e., 863 rows and 283 columns 
```

```python
# so now we are ready with our training data
# then, we have to build our LSTM model's  architecture and train our model.
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input
```

```python
# build our model 

model = Sequential()

# Create an explicit Input layer to "build" the model immediately
model.add(Input(shape=(56,)))

# creating 'embedding_layer'
# In our o/p dense vector, there will be 100 numbers
# b/c each row contains 56 words
model.add(Embedding(283, 100, input_length=56))    

# now add our LSTM layer
model.add(LSTM(150))    # how many nodes inside each lstm gates

# finally add dense layer
model.add(Dense(283, activation='softmax'))       # in this 'no. of nodes' is equal to 'words present in the vocabulary' 
# for every word we get a probability, and one which have highest probability, will be our 'output' finally.
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# we take (loss='categorical_crossentropy') b/c it is a multi-class problem.
```

```python
model.summary()
```

```python
# train the model 

model.fit(X, y, epochs=100) 
```

```python
# we can see the 'accuracy' for our 'training_data' is 95%, which is okay.
```

```python
# Now, we have to predict next word for a given word

text = "mail"       # let's say we choose this word

# first we have to tokenize it
token_text = tokenizer.texts_to_sequences([text])[0]

# And then we have to apply padding on it 
padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')

# finally do prediction for next word
model.predict(padded_token_text)      

# so we get 283 outputs (that shows we get probability across each 'words' which is present in our 'training_data')
# And we have to find that word which having highest probability
```

```python
import numpy as np

pos = np.argmax(model.predict(padded_token_text))    # It tells us that at whuich position we have the highest probability

# so this shows that word present at 160th index having the highest probabilty as an output.
```

```python
tokenizer.word_index.items()    # so this is our dictionary and we have to find (index_number=160) from this.
```

```python
# for predicting next word :-

for word,index in tokenizer.word_index.items():
    if index == pos:
      print(word)
```

```python
# If we want to predict next 5 words, then-

import time
text = "nlp and deep learning"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
